```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
np.random.seed(42)
```

```python
neural_data = np.random.rand(500, 50)
```

```python
print("Neural data shape:", neural_data.shape)
```

```python
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(neural_data)
```

```python
print("Reduced data shape:", reduced_data.shape)
print("Explained variance ratio:", pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], alpha=0.7, c='b')
plt.title('PCA of Simulated Neural Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.bar(range(1, pca.n_components + 1), pca.explained_variance_ratio_[:pca.n_components], alpha=0.7)
plt.title('Variance Explained by Principal Components')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.grid(True)
plt.show()
```
