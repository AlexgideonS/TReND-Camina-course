```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
W = np.array([[0.6, 0.2, 0.0],
              [0.1, 0.5, 0.3],
              [0.0, 0.3, 0.7]])
```

```python
x = np.array([1.0, 0.0, 0.5])
```

```python
T = 20
activity_history = [x.copy()]
```

```python
for t in range(T):
  x = W @ x
  activity_history.append(x.copy())
activity_history = np.array(activity_history)
```

```python
plt.figure(figsize=(10, 5))
for neuron in range(3):
    plt.plot(activity_history[:, neuron], label=f'Neuron {neuron+1}')
plt.xlabel('Time step')
plt.ylabel('Activity level')
plt.title('Recurrent Neural Circuit Simulation')
plt.legend()
plt.grid(True)
plt.show()
```

```python
np.random.seed(0)
neural_data = np.random.rand(50, 100)
```

```python
data_centered = neural_data - np.mean(neural_data, axis=0)
```

```python
U, S, VT = np.linalg.svd(data_centered, full_matrices=False)
```

```python
reduced_data = data_centered @ VT.T[:, :2]
```

```python
plt.figure(figsize=(8,6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], alpha=0.7, c='blue')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Neural Activity (100 neurons)')
plt.grid(True)
plt.show()
```
